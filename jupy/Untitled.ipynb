```python
import numpy as np
import matplotlib.pyplot as plt

e=2.718281828459

# Define the function
def f(x):
    return np.exp(x)


# Generate x values
x = np.linspace(0, 5, 100)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label='$e^x$')
plt.plot(x, f(x)/e, label='$e^{x-1}$')
plt.plot(x, f(x)/e**2, label='$e^{x-2}$')
plt.plot(x, f(x)/e**3, label='$e^{x-3}$')
plt.plot(x, f(x)/e**4, label='$e^{x-4}$')
plt.axvline(x=1.6,color="Black")
plt.text(0.4,1.8,"$e^x$")
plt.text(1.2,1.8,"$e^{x-1}$")
plt.text(2.2,1.8,"$e^{x-2}$")
plt.text(3.2,1.8,"$e^{x-3}$")
plt.text(4.2,1.8,"$e^{x-4}$")

plt.xlabel('acceleration')
plt.ylabel('mass ejected')
plt.title('Graph of $e^x$')
plt.grid(True)
plt.legend()

# Define the scaling factor for the y-axis
scale_factor = 0.04  # Adjust this value as needed

# Calculate the limits of the y-axis
y_min, y_max = plt.ylim()
new_y_min = y_min * scale_factor
new_y_max = y_max * scale_factor

# Set the limits of the y-axis with the scaled values
plt.ylim(new_y_min, new_y_max)

plt.show()
```
