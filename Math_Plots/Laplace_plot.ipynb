```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the complex function
def complex_function(z):
    return z**2 + np.sin(z)

def two_pole(z):
    denom = (z + 1) * (z + 2)
    return 1 / denom

# Create a grid in the complex plane
x = np.linspace(-3, 0, 50)  # Real part
y = np.linspace(-1, 1, 50)  # Imaginary part
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Evaluate the function on the grid
W = two_pole(Z)

# Extract magnitude and cap it at 10
magnitude = np.abs(W)
magnitude_capped = np.minimum(magnitude, 10)  # Apply height cap
phase = np.angle(W)

# Create the 3D plot for the magnitude
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, magnitude_capped, cmap='viridis', edgecolor='k')

# Add color bar for magnitude
fig.colorbar(surf, ax=ax, shrink=1, aspect=10, label='Magnitude (Capped at 10)')

# Labels and title
ax.set_title("3D Plot of |f(z)| with Height Cap at 10")
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_zlabel("|f(z)|")

plt.show()
```
